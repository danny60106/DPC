```python

import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np

from dpca import DensityPeakCluster

# file name
file = "jain"

# load data

data = pd.read_csv(r"data/coordinationData/%s.csv" % file, header=None)

# dpca model
# plot decision graph to set params `density_threshold`, `distance_threshold`.
dpca = DensityPeakCluster(dc= 0.7650391198643011,density_threshold=5, distance_threshold=7, anormal=False)

# fit model
dpca.fit(data.iloc[:, [0, 1]])

# print predict label
print(dpca.labels_)

# plot cluster
dpca.plot("all", title=file, save_path="data/result")

print(dpca.dc)
```

```python
def local_density(max_id, distances, dc, guass=True, cutoff=False):
    
	assert guass and cutoff == False and guass or cutoff == True
	guass_func = lambda dij, dc : math.exp(- (dij / dc) ** 2)
	cutoff_func = lambda dij, dc: 1 if dij < dc else 0
	func = guass and guass_func or cutoff_func
	rho = [0] * max_id
	for i in range(max_id):
		for j in range(i + 1, max_id):
			temp = func(distances[(i, j)],dc)
			rho[i] += temp
			rho[j] += temp
	return np.array(rho, np.float32)
```

```python
def calcGini(dc,guass=True, exponent=2):
    rho = local_density(max_id, distances, dc, guass=True, cutoff=False)
    #print(rho[:10])
    #print(rho[0])
    Z = np.sum(rho)    
    gini = 1 - np.sum(np.power(rho/Z,2))
    print("Z=", Z)
    print("gini=", gini)
    #print(np.sum(rho/Z))
    #print(np.sum(np.power(rho/Z,2)))
    return gini
```

```python
def findDcByMinGini(min_dis):
    dclis=[]
    ginilis=[]
    for i in range(1, 20, 1):
        dc = min_dis*i
        gini = calcGini(dc, guass=True, exponent=2)
        dclis += [dc]
        ginilis += [gini]
        print(i, '\t', dc, '\t', gini)
    plt.plot(dclis,ginilis)
```

```python
def findKNNandDis(distances,knn,per):
    sort= sorted(distances.items(), key=lambda x:x[1])
    sort_dict=dict(sort)
    k=[]
    for j in range (0,max_id):
        for i in range (0,max_id):
            if i != j:
                k += [sort_dict[i,j]]           
    k_array=np.array(k).reshape(max_id,max_id-1)
    c=np.sort(k_array,axis=1)
    sum_knn=[]
    for i in range(0,max_id):
        sum_knn += [(c[i, knn-1])]
    knn_sort=sorted(sum_knn)
    mid=max_id*per
    knn_dc=sum(knn_sort[0:int(mid)])/mid
    f_dis={}
    for i in range (0,max_id):
        for j in range (0,max_id):
            if i != j and distances[i,j] <= knn_dc:
                f_dis.update({(i,j): distances[i,j]})
            
    l = np.zeros((max_id,max_id))
    c = []
    s = []
    f = []
    for i in range (0,max_id+1):
        for j in range (i+1,max_id+1):
            if i!= j and f_dis.__contains__((i,j)):
               c.append((i,j))
               s.append(i)
               f.append(j)
    return c,s,f
```

```python
class Graph:
 
    # init function to declare class variables
    def __init__(self, V):
        self.V = V
        self.adj = [[] for i in range(V)]
 
    def DFSUtil(self, temp, v, visited):
 
        # Mark the current vertex as visited
        visited[v] = True
 
        # Store the vertex to list
        temp.append(v)
 
        # Repeat for all vertices adjacent
        # to this vertex v
        for i in self.adj[v]:
            if visited[i] == False:
 
                # Update the list
                temp = self.DFSUtil(temp, i, visited)
        return temp
 
    # method to add an undirected edge
    def addEdge(self, v, w):
        self.adj[v].append(w)
        self.adj[w].append(v)
 
    # Method to retrieve connected components
    # in an undirected graph
    def connectedComponents(self):
        visited = []
        cc = []
        for i in range(self.V):
            visited.append(False)
        for v in range(self.V):
            if visited[v] == False:
                temp = []
                cc.append(self.DFSUtil(temp, v, visited))
        return cc
 
 
# Driver Code
if __name__ == "__main__":
 
    # Create a graph given in the above diagram
    # 5 vertices numbered from 0 to 4
    g = Graph(max_id)
    for i, j in zip(s, f):
        g.addEdge(i,j)

    
    cc = g.connectedComponents()
    print(cc)
```

```python
import csv
def readDataFromFile(filename, normalized=False):
    Y = []
    with open(filename, newline='') as csvfile:
        #  CSV  dictionary        
        myCsv = csv.reader(csvfile)        
        for row in myCsv:
            # print("xxxxxxxxx", row[0], row[1], row[2])
            Y.append(np.array([row[0], row[1]], dtype=np.float32))
    Y= np.array(Y)
    
    return Y
```

```python
def plot_answer_graph(X, labels_true):
    fig=plt.figure(figsize=(16, 16), dpi= 80, facecolor='w', edgecolor='k')

    unique_labels = set(labels_true)   #  (-1)

    print(unique_labels)

    colors = [plt.cm.Spectral(each)
              for each in np.linspace(0, 1, len(unique_labels))]  

    for k, col in zip(unique_labels, colors):   # 
        if k == -1:
            # Black used for noise.
            col = [0, 0, 0, 1]

        class_member_mask = (labels_true == k)  # k

        xy = X[class_member_mask]  # k 
        plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
                 markeredgecolor='k', markersize=15)  # 
```
